```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# 데이터 로드
csv_path = 'Life Expectancy.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# 데이터 확인
df.info()
```

```python
df.shape
```

```python
# check for duplicate values
df.duplicated().value_counts()
```

```python
# 결측값 처리 (간단한 결측값 제거로 처리)
df = df.dropna()
```

```python
df.describe()
```

```python
# 히트맵 그리기
df_num = df.select_dtypes(include=[int, float, bool])
plt.figure(figsize=(18,18))  # 크기 조정
sns.heatmap(df_num.corr(), annot=True, fmt=".2f", cmap='coolwarm', cbar=True, annot_kws={"size": 10})  # 색상 팔레트 변경 및 숫자 형식 조정
plt.xticks(rotation=45, ha='right', fontsize=10)  # x축 레이블 회전 및 폰트 크기 조정
plt.yticks(fontsize=10)  # y축 폰트 크기 조정
plt.title('Correlation Heatmap', fontsize=16)  # 제목 추가
plt.show()
```

```python
# GDP과 Total Expenditure의 관계 시각화 (이건 수립한 가설과는 관계없지만 그냥 해봄)
df_grouped = df.groupby('GDP', as_index=False)['percentage expenditure'].mean()
fig = px.bar(df_grouped, x='GDP', y='percentage expenditure', template='presentation', title='<b>percentage expenditure based on GDP</b>')
fig.show()
```

```python
# GDP의 분포 시각화 - 히스토그램(아래 Scatterplot에서 GDP 구간을 설정해주기 위함)
plt.figure(figsize=(12, 6))
sns.histplot(df['GDP'], bins=30, kde=False)
plt.title('Distribution of GDP')
plt.xlabel('GDP')
plt.ylabel('Frequency')
plt.show()
```

```python
# 주요 지표와 기대 수명 간의 관계 시각화
plt.figure(figsize=(16, 12))

# GDP와 기대 수명
plt.subplot(2, 2, 1)
sns.scatterplot(x='GDP', y='Life expectancy ', data=df)
plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.xlim(0, 20000)  # x축 범위 설정

# ICOR와 기대 수명
plt.subplot(2, 2, 2)
sns.scatterplot(x='Income composition of resources', y='Life expectancy ', data=df)
plt.title('Income Composition of Resources vs Life Expectancy')
plt.xlabel('Income Composition of Resources')
plt.ylabel('Life Expectancy')

# Schooling과 기대 수명
plt.subplot(2, 2, 3)
sns.scatterplot(x='Schooling', y='Life expectancy ', data=df)
plt.title('Schooling vs Life Expectancy')
plt.xlabel('Schooling')
plt.ylabel('Life Expectancy')

# 성인 사망률과 기대 수명
plt.subplot(2, 2, 4)
sns.scatterplot(x='Adult Mortality', y='Life expectancy ', data=df)
plt.title('Adult Mortality vs Life Expectancy')
plt.xlabel('Adult Mortality')
plt.ylabel('Life Expectancy')

plt.tight_layout()
plt.show()
```

```python
# 연도별 평균 기대 수명 계산
df_grouped = df.groupby('Year')['Life expectancy '].mean().reset_index()

# 기대 수명 추이 시각화
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_grouped, x='Year', y='Life expectancy ', marker='o')
plt.title('Trends in Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
plt.show()
```
